## Business Background:

First Bank is a small retail bank in Portugal that provided Big Data Energy, our company, with customer data from a publicly accessible banking dataset\. First Bank has noticed a growing number of customers maintaining balances close to or below zero\. The banking dataset contains demographic information about customers, the financial products they hold, and the amounts in each of their bank accounts\. First Bank has sent Big Data Energy a large sample of its customer data for further analysis\. This issue is causing the bank to lose money and preventing its clients from accessing better products, such as term deposits and loans\. First Bank has observed that many customer accounts maintain balances that are very close to or below $0, which may indicate overdraft risk and financial instability among certain customer segments\.

To gain additional knowledge about this problem, First Bank hired Big Data Energy to collect data on all of First Bank's customers and examine patterns in account balances and the use of First Bank's financial products\. The purpose of this analysis was to determine whether specific attributes of certain customer demographics or customer actions could be linked to smaller account balances; then to develop targeted financial products or provide financial counseling for those customer segments identified as potentially having small account balances\. Understanding which customers are more likely to maintain balances near or below $1000 can help the bank design strategies that reduce overdraft risk and improve overall customer financial health\.

The focus for this project will be limited to a subset of variables that best illustrate the banks' questions concerning the characteristics of their customers and the factors related to the use of financial products by those customers\. Those variables included Age, Balance, Loan, Housing, Default, and Y \(term deposit subscription\)\. The Age variable will be analyzed to see if particular age groups are more likely than others to have low or negative account balances\. The Balance variable serves as a proxy for a customer's financial stability relative to their account\. The Loan and Housing variables are being examined to determine whether customers who have taken advantage of other financial products offered by the bank \(i\.e\., personal loan, housing loan\) exhibit behaviors that would be expected to affect their account balance\. If a customer has a history of failing to repay loans \(i\.e\., a default\), it is reasonable to assume that the customer may have a lower account balance due to the impact of credit problems on a customer's ability to obtain new credit\. Together, these variables allow the bank to explore whether financial behavior, demographics, or product usage are associated with customers maintaining balances close to or below $1000\.

First Bank is seeking to use these variables to understand why some of its customers have lower account balances, whether some of its customers with lower account balances are currently using some type of financial service provided by First Bank, and how the findings of this analysis can lead to the development of targeted financial products that will assist customers in improving their account balances\. Ultimately, First Bank wants to convert customer banking data into actionable insights that will enable it to build stronger relationships with its customers, help its customers make better financial decisions, and facilitate more informed decision\-making at the bank\.

### Data Manipulation:

In this section, we have organized our dataset and developed a number of additional variables to prepare our data for both analysis and predictive model development\. The categorical variables of loan status, credit default, housing loan, and the target variable were transformed from "yes" and "no" responses into binary \(0 and 1\) representations for use within the statistical models\. Customer ages were then binned by 10 years to reduce complexity in analysis and potentially highlight trends that may exist in terms of age and financial behavior\. Account balances were also placed into various categories of balance ranges, along with the creation of a binary variable to determine which customers had low account balances \(less than $1,000\)\. These modifications will assist us in making the dataset easier to analyze and allow us to identify potential patterns associated with low account balances\.

After adding the new variables to the original dataset, we built a cleaned dataset containing only the variables needed for our analysis\. A key benefit of this cleaning process is that it removes irrelevant variables and organizes the data for easier analysis or modeling\. We selected to include variables such as age groups, balance bins, loan status, housing loans, and a low account balance indicator because they are the features of the data most likely to be useful to us in answering our business\-related questions\.

```
#Organizing data and adding new fields to the bank data set

#Changing any yes or no values to 0 and 1

#updating loan from yes and no to 1 and 0
mapping_dict = {'yes': 1, 'no': 0}
bank['loan_binary'] = bank['loan'].map(mapping_dict)

#updating default from yes and no to 1 and 0
mapping_dict = {'yes': 1, 'no': 0}
bank['default_binary'] = bank['default'].map(mapping_dict)

#update housing from yes and no to 1 and 0
mapping_dict = {'yes': 1, 'no': 0}
bank['housing_binary'] = bank['housing'].map(mapping_dict)

#Create age groups bin, ages are being binned by 10 year increments and labeling accordingly
bank['age_group'] = pd.cut(bank['age'], bins=[17, 28, 38, 48, 58, 68, 78, 88, 98], 
labels=['18–28', '29–38', '39–48', '49–58', '59–68', '69-78', '79-88', '89+'])

#Creating age groups by number instead of range of lables
bank['age_group_bins'] = pd.cut(bank['age'], bins=[17, 28, 38, 48, 58, 68, 78, 88, 98], 
labels=[1, 2, 3, 4, 5, 6, 7, 8]).astype(int)


#Create balance bin, creating bins to organize account balance in 5000 increments starting with a -$10,000 balance and moving up in $5000 increments. 
bank['balance_bin'] = pd.cut(bank["balance"], [-10000,-5000, 0, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000, 95000, 100000, 150000])
bank.groupby("balance_bin").size().astype(int)


#Group cusomters by account balance that is greater or less than $1000 to help narrow down customer account balances
bank['low_account_balance'] = pd.cut(bank["balance"], bins = [-10000, 1000, 1500000],
labels = ['<1000', '+1000'])

#Creating 0 and 1 bins for account balances
bank['low_account_balance_binary'] = pd.cut(bank["balance"], bins = [-10000, 1000, 1500000],
labels = [0, 1]).astype(int)

#creating a field that just has blance below 0
bank['below_zero'] = (bank['balance'] < 0).astype(int)
print(bank.head())
```

```
bank_clean = bank[[
    'loan', 
    'loan_binary',
    'balance',
    'balance_bin', 
    'age', 
    'default',
    'default_binary', 
    'housing', 
    'housing_binary',
    'age_group',
    'age_group_bins', 
    'low_account_balance',
    'low_account_balance_binary',
    'below_zero']]
print(bank_clean.info())
```

## Training and Test Dataset:

The bank data set has been split out into a test dataset, which consists of 25% of the training data\. Before building the predictive model, several preprocessing steps were performed to prepare the dataset for analysis\. These included selecting the relevant variables identified during data exploration, ensuring data types were appropriate for modeling, and preparing the target variable for classification analysis\. Categorical variables such as Loan, Housing, and Default were encoded so that they could be used in the logistic regression model\. The dataset was also cleaned to ensure there were no missing values or incorrect data types that could affect the analysis\. The analysis will check for correlation between the target y of low account balance and the other fields\.

```
# Splitting data sets into a train and test data set and copying into a new dataframe 

X_train = bank_clean[['loan_binary', 'default_binary', 'housing_binary', 'age_group_bins']].copy()
y_train = bank_clean['low_account_balance_binary'].copy()

X = bank_clean[['loan_binary', 'default_binary', 'housing_binary', 'age_group_bins']].copy()
y = bank_clean['low_account_balance_binary'].copy()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)


print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)

print(X_train.dtypes)
```

```
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_test, y_test)
print("Intercept:", round(logreg.intercept_[0], 2))
```

```
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print(cnf_matrix)
```

```
class_names= [0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

#creating the heat map
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap='YlGnBu', fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y = 1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

This heat map summarizes the classification performance of the logistic regression model on the test dataset\. The matrix compares the model's predicted outcomes with the actual outcomes for whether customers have a low account balance \(less than $1000\)\. The top\-left cell \(7,579\) represents true positives, where the model correctly predicted that customers would have an account balance below $1000\. The bottom\-left cell \(3,582\) represents false positives, where the model predicted that customers did not have an account balance below $1000\. The upper right corner of 66 indicates the false negatives, and the lower right of 76 indicates the true negatives\.

```
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("ROC AUC:", roc_auc)
```

To evaluate the performance of the logistic regression model, several classification metrics were calculated, including accuracy, precision, recall, and ROC AUC\. The model achieved an accuracy of approximately 0\.68, meaning that about 68% of the predictions made by the model correctly classified whether a customer had an account balance of $1000 or less\. The precision score of 0\.68 indicates that when the model predicts a customer will have a low account balance, it is correct about 68% of the time\. The recall score is very high at approximately 0\.99, meaning the model successfully identifies nearly all customers who have low account balances\. However, the ROC AUC score of about 0\.51 suggests that the model does not strongly distinguish between customers who do and those who do not, as this value is only slightly better than random guessing\. Overall, while the model is effective at identifying most of the positive cases \(high recall\), the relatively low ROC AUC indicates that additional predictors or model improvements may be needed to improve overall predictive performance\.

```
from sklearn import metrics
import matplotlib.pyplot as plt

y_pred_proba = logreg.predict_proba(X_test)[::, 1]

fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)

auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.title("ROC Curve Results")
plt.plot(fpr, tpr, label="data 1, auc = " + str(auc))
plt.legend(loc=4)
plt.show()
```

ROC \(Receiver Operating Characteristic\) curve evaluates the performance and accuracy of the logistic regression model that is used to predict whether a customer will have an amount balance of less than $1000\. The curve illustrates the trade\-off between the true positive rate \(sensitivity\) and the false positive rate across different classification thresholds\. The area under the curve \(AUC\) is approximately 0\.60, indicating that the model performs slightly better than random guessing but does not have strong predictive power\. While the model can identify some patterns in the data, the relatively modest AUC suggests that the current set of predictors may not fully capture the factors influencing whether customers have a low account balance\. For First Bank, these results suggest that additional variables or behavioral indicators may be needed to improve the model's ability to accurately predict account balances\.
